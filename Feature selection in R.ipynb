```r
library(dplyr)
# Read in CSV file with features, produced in Python notebook
feats <- read.csv('all_feats.csv')

# Get rid of entire sample if any NA values
feats <- na.omit(feats)

# Data frame with only HRV features
#feats_HR <- select(feats, HRV_RMSSD, HRV_MeanNN, HRV_SDNN, HRV_SDSD, HRV_CVNN, HRV_CVSD, HRV_MedianNN, HRV_MadNN, HRV_MCVNN, HRV_IQRNN, HRV_pNN50, HRV_pNN20, HRV_TINN, HRV_HTI, PR_mean, PR_sd, ST_mean, ST_sd, PT_mean, PT_sd, RS_mean, RS_sd, RT_mean, RT_sd, PS_mean, PS_sd,stress)
feats_HR <- select(feats, PR_mean, PR_sd, ST_mean, ST_sd, PT_mean, PT_sd, RS_mean, RS_sd, RT_mean, RT_sd, PS_mean, PS_sd,stress)

# Produce linear model with all features
model <- lm(stress ~ ., data = feats_HR)

# Print model summary
summary(model)

```

```r
library(MASS)
stepAIC(model, 
        direction = c("both"),
        trace = TRUE)
```

```r
# Produce linear model with all features
model_trad <- lm(stress ~ HRV_RMSSD + HRV_MeanNN + HRV_SDNN + HRV_SDSD + HRV_CVNN + HRV_CVSD + HRV_MedianNN + HRV_MadNN + HRV_MCVNN
            + HRV_IQRNN + HRV_pNN50 + HRV_pNN20 + HRV_TINN + HRV_HTI, data = feats_HR)
summary(model_trad)
```

```r
library(MASS)
stepAIC(model_trad, 
        direction = c("both"),
        trace = TRUE)
```
